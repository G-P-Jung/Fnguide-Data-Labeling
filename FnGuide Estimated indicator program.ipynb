# 기업 추정지표 데이터 전처리 프로그램
- 데이터 파일 위치는 본인의 저장위치에 맞게 변경하여 진행하시면 됩니다
- file_name() -> 전처리를 원하는 파일 입력 함수
- 추정지표_col_kor() -> 기업_추정_{}_표본.csv data columns 한글 변경 및 파일 내보내기 함수
    - 기업_추정_{}_표본_col_kor.csv 데이터로 저장됨
- 추정지표_data_kor() -> 기업_추정_{}_표본_col_kor.csv data 기간구분코드, 기업코드, 계정코드 변경 및 파일 내보내기 함수
    - 기업_추정_{}_표본_data_kor.csv 데이터로 저장됨

```python
import numpy as np
import pandas as pd
```

```python
def file_name():
    data_name_list = ['재무재표', '재무비율', '투자지표']
    while(True):
        data_name = input("전처리 할 데이터 이름을 입력해주세요(재무재표, 재무비율, 투자지표 중 하나 입력) : ")
        if data_name in data_name_list:
            print("기업_추정_{}_표본.csv 파일 전처리를 진행하겠습니다. 잠시 기다려주세요.....".format(data_name))
            return data_name
            break
        else :
            print("다시 입력해주세요")
```

```python
def 추정지표_col_kor(data_name):
    df = pd.read_csv("/Users/jeong-gwangpil/에프엔가이드 데이터 전처리/Estimated indicator data/기업_추정_{}_표본.csv".format(data_name))
    df.columns = [['기업코드', '생성일자', '기간구분코드', '재무년월', '계정코드', '데이터값']]
    df.to_csv("/Users/jeong-gwangpil/에프엔가이드 데이터 전처리/추정지표_col_kor/기업_추정_{}_표본_col_kor.csv".format(data_name), encoding = 'utf-8-sig', index = False)
    
    print()
    print()
    print("기업_추정_{}_표본.csv columns 한글화를 완료했습니다".format(data_name))
    print("파일명은 기업_추정_{}_표본_col_kor.csv로 변경되어 저장했습니다".format(data_name))
    print("저장위치는 '/Users/jeong-gwangpil/에프엔가이드 데이터 전처리/추정지표_col_kor' 입니다")
```

```python
def 추정지표_data_kor(data_name):
    df = pd.read_csv("/Users/jeong-gwangpil/에프엔가이드 데이터 전처리/추정지표_col_kor/기업_추정_{}_표본_col_kor.csv".format(data_name))
    df1 = pd.read_csv("/Users/jeong-gwangpil/Desktop/빅데이터 공모전/에프엔가이드/use data/기업의 개요/기업코드.csv")
    df2 = pd.read_csv("/Users/jeong-gwangpil/Desktop/빅데이터 공모전/에프엔가이드/use data/추정계정/추정계정.csv")
    
    df['기간구분코드'].replace({1: '누적', 2 : '순액', 3 : 'TTM', 12 : '발표순액', 31 : '연결 누적', 32 : '연결 순액', 33 : '연결 TTM', 42 : '연결 발표순액'}, inplace = True)
    # 기업코드 파일의 기업코드 column list화
    a = df1['기업코드']
    a_val = a.values
    a_list = a_val.tolist()

    # 기업코드 파일의 국문기업명 column list화
    b = df1['국문기업명']
    b_val = b.values
    b_list = b_val.tolist()
    
    for i in range(len(df)):
        if df['기업코드'][i] in a_list:
            df['기업코드'].replace({df['기업코드'][i] : b_list[a_list.index(df['기업코드'][i])]}, inplace = True)
        
    # 추정계정 파일의 계정코드 column list화
    c = df2['계정코드']
    c_val = c.values
    c_list = c_val.tolist()

    # 추정계정 파일의 국문명 column list화
    d = df2['국문명']
    d_val = d.values
    d_list = d_val.tolist()
    
    for i in range(len(df)):
        if df['계정코드'][i] in c_list:
            df['계정코드'].replace({df['계정코드'][i] : d_list[c_list.index(df['계정코드'][i])]}, inplace = True)
#         else :
#             df['계정코드'].replace({df['계정코드'][i] : 0}, inplace = True)
    
    df.to_csv("/Users/jeong-gwangpil/에프엔가이드 데이터 전처리/추정지표_data_kor/기업_추정_{}_표본_data_kor.csv".format(data_name), encoding = 'utf-8-sig', index = False)
    
    print()
    print()
    print("기업_추정_{}_표본_col_data.csv columns 한글화를 완료했습니다".format(data_name))
    print("파일명은 기업_추정_{}_표본_data_kor.csv로 변경되어 저장했습니다".format(data_name))
    print("저장위치는 '/Users/jeong-gwangpil/에프엔가이드 데이터 전처리/추정지표_data_kor' 입니다")
    
    return df
```
